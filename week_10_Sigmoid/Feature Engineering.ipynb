```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from imperio import FrequencyImputationTransformer
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
df = pd.read_csv('heart.csv')
df = pd.DataFrame(df)
X = df.drop('target', axis=1)
Y = df['target']
X_train,X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.33, random_state = 10)
print(X_test.head(10))
pipe = Pipeline([('tran1',StandardScaler()),('tran2',FrequencyImputationTransformer(index = [2],min_int_freq = 2))])
# X_train.reshape(-1,1)
# Y_train.reshape(-1,1)
pipe.fit(X_train,Y_train)
output = pipe.transform(X_test)
# score = pipe.score(X_test, Y_test)
# print('Accuracy:', score)
output = pd.DataFrame(output, columns= list(X))
print(output['chest pain type'])
```
