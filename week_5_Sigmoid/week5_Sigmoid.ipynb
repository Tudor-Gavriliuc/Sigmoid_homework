# Train a from-scratch implementation of Linear Regression on the train sub-set.

```python
import numpy as np
#print(y_train.values.reshape(-1,1).shape)
class MultipleLinearRegression:
    def __init__(self, alpha=0.00000000000000001, n_iterations=1000):
        self.alpha = alpha
        self.n_iterations = n_iterations

    def fit(self, X, y):
        # Add a column of ones for the intercept term
        ones = np.ones((X.shape[0], 1))
        X = np.concatenate((ones, X), axis=1)

        # Initialize the weights
        self.weights = np.ones((X.shape[1], 1))

        # Gradient descent
        for i in range(self.n_iterations):
            y_pred = np.dot(X, self.weights)
            error = y_pred - y
            gradient = np.dot(X.transpose(), error) / y.size
            self.weights = self.weights -  self.alpha * gradient

    def predict(self, X):
        # Add a column of ones for the intercept term
        ones = np.ones((X.shape[0], 1))
        X = np.concatenate((ones, X), axis=1)
        print(np.dot(X,self.weights))
        # Predict using the learned weights
        return np.dot(X, self.weights)
    def r2_score(y_true, y_pred):
        ss_res = np.sum((y_true - y_pred) ** 2)
        ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
        r2 = 1 - (ss_res / ss_tot)
        print(r2)
        return r2
#y_train = y_train.reshape(-1,1)
nem =  MultipleLinearRegression()
nem.fit(X_train,y_train)
obj = nem.predict(X_test)
```
